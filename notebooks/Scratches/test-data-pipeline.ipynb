```python
def data_generator():
    with open('../../data/processed/train/strings_X.txt', 'r') as f:
        X = f.readlines()
    with open('../../data/processed/train/strings_Y.txt', 'r') as f:
        Y = f.readlines()
    X = [x.strip('\n') for x in X]
    Y = [y.strip('\n') for y in Y]
    ws = WhiteSpaceTokenizer()
    padder = Padder('<pad>')
    pipe = Pipeline(steps=[('ws', ws),('pad', padder)])
    X_processed = pipe.transform(X)
    
    word2vec = Word2Vec()
    model = KeyedVectors.load_word2vec_format('../../data/processed/short-vectors.bin.gz', binary=True)
    word2vec.set_model(model)
    
    for i, x in enumerate(X_processed):
        yield (word2vec.map(x), Y[i])
```

```python
class training_data_generator():
    def __init__(self, data_dir, input_fpath, output_fpath):
        self.data_dir = data_dir
        self.input_fpath = input_fpath
        self.output_fpath = output_fpath
#         self.vocab = Vocab()
        self.load_txt()
        self.i = 0
        
    def load_txt(self):
        with open('../../data/processed/train/strings_X.txt') as f:
            X = f.readlines()
        with open('../../data/processed/train/strings_Y.txt') as f:
            Y = f.readlines()
    
    def __iter__(self):
        return self
    
    def __next__(self):
        return self.next()
    
    def next(self):
        self.i+=1
        return X[self.i]
    
    def get(self):
        for i, x in enumerate(X):
            yield x
```
